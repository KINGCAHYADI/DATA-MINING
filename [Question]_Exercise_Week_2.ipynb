```python
# h
import matplotlib.pyplot as plt

# Menghitung jumlah hero per peran
role_counts = mlbb_df_cleaned['role'].value_counts()

# Menyiapkan data untuk pie chart
labels = role_counts.index
sizes = role_counts.values

# Membuat pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Menambahkan judul
plt.title('PERCENTAGE OF HEROES BY ROLES')

# Menampilkan pie chart
plt.show()
```

7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Encode the 'role' column
encoder = LabelEncoder()
y = encoder.fit_transform(mlbb_df['role'])

# Separate features and target variable
X = mlbb_df[['defense_overall', 'offense_overall', 'skill_effect_overall', 'difficulty_overall', 
         'movement_spd', 'magic_defense', 'mana', 'hp_regen', 'physical_atk', 
         'physical_defense', 'hp', 'attack_speed', 'mana_regen']]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)

# Create a Decision Tree Classifier
model = DecisionTreeClassifier()

# Train the model
model.fit(X_train,y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Predict the role for the given character
character = [[6.0,7.0,4,3,250,9,380,37,119,24,2689,0.75,16]]
predicted_role = model.predict(character)

# Get the predicted role (dicoding)
predicted_role_name = encoder.inverse_transform(predicted_role)
print("Predicted Role:", predicted_role_name[0])
```

8. Why is CSV one of the most commonly used file formats in the field of data science? <br>

```python
# 8 CSV (Comma-Separated Values) merupakan salah satu format file yang paling umum digunakan dalam ilmu data karena beberapa alasan. Pertama, format ini sangat sederhana dan berbasis teks, di mana setiap baris mewakili satu record dan kolom dipisahkan dengan koma, sehingga mudah dibaca dan ditulis. Kedua, CSV didukung oleh hampir semua bahasa pemrograman dan aplikasi analisis data, termasuk Excel, memungkinkan pertukaran data yang lancar antar sistem. Selain itu, file CSV bersifat mudah dibaca manusia, efisien dalam penyimpanan, dan fleksibel dalam menangani berbagai jenis data. Format ini juga dapat dihasilkan dengan mudah oleh berbagai alat, menjadikannya pilihan yang ideal untuk proses pengumpulan, pemrosesan, dan berbagi dataset dalam proyek ilmu data.
```
